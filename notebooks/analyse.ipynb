```python
import seaborn as sns

# Create the plot
plt.figure(figsize=(12, 6))
ax = sns.barplot(data=df, y="Drones Travel Distance", x="strategy")

bar_min = df.groupby("strategy")["Drones Travel Distance"].mean().min()
bar_max = df.groupby("strategy")["Drones Travel Distance"].mean().max()
# Set the y-axis limits
y_min = bar_min * 0.99
y_max = bar_max * 1.01
ax.set_ylim(y_min, y_max)

plt.title("Total Drone Travel Distance by Strategy", fontsize=16)
plt.xlabel("Strategy", fontsize=12)
plt.ylabel("Total Drone Travel Distance", fontsize=12)
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
import math

import matplotlib.pyplot as plt
import seaborn as sns

# List of metrics you want to plot
metrics = [
    "Drones Travel Distance",
    "Drones Active Time",
    "Total Distribution Time",
    "Drone Utilization",
    "avg_span_2",
    "avg_span_3",
    "avg_span_4",
    "avg_nodropoffs_2",
    "avg_nodropoffs_3",
    "avg_nodropoffs_4",
    "median_span_2",
    "median_span_3",
    "median_span_4",
    "median_nodropoffs_2",
    "median_nodropoffs_3",
    "median_nodropoffs_4",
    "The Number of Battery Swaps",
    "Total Package Delivery Time",
    "Total Customer Delivery Time",
    "Average Customer Delivery Time",
    "Average Package Delivery Time",
    "Median Package Delivery Time",
    "Median Customer Delivery Time",
]


# Calculate the grid size
n_metrics = len(metrics)
n_cols = 3  # You can adjust this to change the number of columns
n_rows = math.ceil(n_metrics / n_cols)

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(6 * n_cols, 5 * n_rows))
axes = axes.flatten()  # Flatten the 2D array of axes for easier indexing

for i, metric in enumerate(metrics):
    ax = axes[i]

    # Create the plot
    sns.barplot(data=df, y=metric, x="strategy", ax=ax)

    # Calculate min and max
    bar_min = df.groupby("strategy")[metric].mean().min()
    bar_max = df.groupby("strategy")[metric].mean().max()

    # Set the y-axis limits
    y_min = bar_min * 0.99
    y_max = bar_max * 1.01
    ax.set_ylim(y_min, y_max)

    # Set titles and labels
    ax.set_title(f"{metric} by Strategy", fontsize=12)
    ax.set_xlabel("Strategy", fontsize=8)
    ax.set_ylabel(metric, fontsize=8)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")

# Remove any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
