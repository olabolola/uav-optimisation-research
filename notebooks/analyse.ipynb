```python
import pandas as pd
import os

# List of CSV file names
csv_files = ['../results/results_10.csv', '../results/results_20.csv', '../results/results_30.csv', '../results/results_40.csv', '../results/results_50.csv']

# Initialize an empty list to store the DataFrames
dfs = []

# Iterate over the CSV files
for file in csv_files:
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file)
    
    # Extract the number of customers per cluster from the file name
    no_customers_per_cluster = int(os.path.splitext(file)[0].split('_')[-1])
    
    # Add the 'no_customers_per_cluster' column to the DataFrame
    df['no_customers_per_cluster'] = no_customers_per_cluster
    
    # Append the DataFrame to the list
    dfs.append(df)

# Concatenate all the DataFrames vertically
combined_df = pd.concat(dfs, ignore_index=True)

# Define a function to calculate the number of clusters based on customers per cluster and total customers
def calculate_no_clusters(row):
    customers_per_cluster = row['no_customers_per_cluster']
    total_customers = row['no_customers']
    
    if customers_per_cluster == 10:
        if total_customers == 50:
            return 5
        elif total_customers == 100:
            return 10
        elif total_customers == 200:
            return 20
        elif total_customers == 500:
            return 50
    elif customers_per_cluster == 20:
        if total_customers == 50:
            return 3
        elif total_customers == 100:
            return 5
        elif total_customers == 200:
            return 10
        elif total_customers == 500:
            return 25
    elif customers_per_cluster == 30:
        if total_customers == 50:
            return 2
        elif total_customers == 100:
            return 4
        elif total_customers == 200:
            return 7
        elif total_customers == 500:
            return 17
    elif customers_per_cluster == 40:
        if total_customers == 50:
            return 2
        elif total_customers == 100:
            return 3
        elif total_customers == 200:
            return 5
        elif total_customers == 500:
            return 13
    elif customers_per_cluster == 50:
        if total_customers == 50:
            return 2
        elif total_customers == 100:
            return 2
        elif total_customers == 200:
            return 4
        elif total_customers == 500:
            return 10

# Apply the calculate_no_clusters function to each row and create the 'no_clusters' column
combined_df['no_clusters'] = combined_df.apply(calculate_no_clusters, axis=1)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv('../results/combined_results.csv', index=False)
```
